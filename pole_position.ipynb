```python
gs = np.array([-3.55, -8.32, -12.44, -15.71, -18.0, -19.59] )
reson = np.array([64.01-12.15j, 50.5-10.3j, 38.17-8.5j, 27.04-6.74j, 17.1-5.0j, 8.34-3.24j, 0.78-1.01j] )

fig, ax = plt.subplots(figsize=(6,3))
ax.scatter(np.real(gs), np.imag(gs), label="Ground state")
ax.scatter(np.real(reson), np.imag(reson), marker='^', label="Resonance state")

plt.xlabel(r"Re$(E)$ [eV]")
plt.ylabel(r"Im$(E)$ [eV]" )
plt.legend()
plt.savefig("./figs/pole_structure.png")
```
