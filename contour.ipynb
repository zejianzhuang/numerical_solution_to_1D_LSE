```julia
using Plots; pythonplot
using LaTeXStrings
```

```julia
include("./src/Dmat.jl")
```

```julia
var = variable()
deter!(E) = Dmat(E, var)[2][1] + Dmat(E, var)[2][2] * 1im
```

```julia
dim = 100
reE = range(0.01, 70, dim)
imE = range(-15, 0, dim)
magn = @. 1 / abs(deter!(reE'+imE*1im) );
magn
```

```julia
cp = contour(reE, imE, magn, levels=200, cbar=false, minorgrid=true)
xlabel!(L"Re($E$) [eV]")
ylabel!(L"Im($E$) [eV]")
savefig("./figs/inverse_abs_det.png")
cp
```
